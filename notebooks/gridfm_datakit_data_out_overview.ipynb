#### Comments

This table is **branch/line-level power-flow output**. Each row corresponds to one transmission line or transformer for a specific scenario.

Key groups of columns:

- **Scenario identifiers**
  - `scenario`: scenario id
  - `load_scenario_idx`: load variation index

- **Branch identity / topology**
  - `idx`: branch id
  - `from_bus`, `to_bus`: buses connected by the branch

- **AC power-flow results**
  - `pf`, `qf`: active/reactive power injected at the **from** side
  - `pt`, `qt`: active/reactive power injected at the **to** side

- **Branch parameters**
  - `r`, `x`, `b`: resistance, reactance, line charging susceptance
  - `tap`, `shift`: transformer tap ratio and phase shift
  - `ang_min`, `ang_max`: angle difference limits
  - `rate_a`: branch thermal rating
  - `br_status`: whether the branch is in service

- **Admittance-related terms**
  - `Yff_r`, `Yff_i`, `Yft_r`, `Yft_i`, `Ytf_r`, `Ytf_i`, `Ytt_r`, `Ytt_i`:
    real and imaginary parts of the branch admittance submatrices

- **DC approximation results**
  - `pf_dc`, `pt_dc`: active power flow from the DC power-flow approximation

#### Bus Data

```python
# Load and inspect bus data
bus_df = pd.read_parquet(bus_path)

print("bus_df shape:", bus_df.shape)
display(bus_df.head(10))
```

#### Comments

This table is **bus-level system state and injections**. Each row corresponds to one bus for a specific scenario.

Key groups of columns:

- **Scenario identifiers**
  - `scenario`
  - `load_scenario_idx`

- **Bus identity**
  - `bus`: bus id

- **Demand / generation at the bus**
  - `Pd`, `Qd`: active and reactive demand
  - `Pg`, `Qg`: active and reactive generation assigned at that bus

- **Voltage solution**
  - `Vm`: voltage magnitude in per-unit
  - `Va`: voltage angle in degrees

- **Bus type indicators**
  - `PQ`, `PV`, `REF`: one-hot style indicators for bus type
    - `PQ=1`: load bus
    - `PV=1`: voltage-controlled generator bus
    - `REF=1`: reference/slack bus

- **Operational limits / shunts**
  - `vn_kv`: nominal voltage base
  - `min_vm_pu`, `max_vm_pu`: voltage magnitude limits
  - `GS`, `BS`: shunt conductance and susceptance

- **DC approximation outputs**
  - `Va_dc`: DC power-flow voltage angle
  - `Pg_dc`: DC active generation

#### Y_bus Data

```python
# Load and inspect Y-bus data
y_bus_df = pd.read_parquet(y_bus_path)

print("y_bus_df shape:", y_bus_df.shape)
display(y_bus_df.head(10))
```

#### Comments

This table is the **bus admittance matrix** in long/tabular form. Each row corresponds to one matrix entry of the network `Y_bus` for a specific scenario.

Key groups of columns:

- **Scenario identifiers**
  - `scenario`
  - `load_scenario_idx`

- **Matrix coordinates**
  - `index1`: row index of the admittance matrix
  - `index2`: column index of the admittance matrix

- **Complex admittance value**
  - `G`: real part of admittance (conductance)
  - `B`: imaginary part of admittance (susceptance)

**Interpretation**: Each row represents one element of

$$
Y_{bus}[index1, index2] = G + jB
$$

- **Diagonal entries** (`index1 == index2`) represent self-admittance of a bus
- **Off-diagonal entries** (`index1 != index2`) represent coupling between two buses

So for example:
- `(0,0)` is the self-admittance of bus 0
- `(0,1)` and `(1,0)` are mutual admittances between buses 0 and 1

#### Constraint violation audit

Checks bus voltage limits, branch thermal/angle limits, and generator active/reactive output limits. Reports both component-row percentages and scenario/case percentages.

```python
import re
import numpy as np
import pandas as pd
from pathlib import Path

# Constraint violation audit across ALL parquet files for case14_ieee only.
#
# Assumes datasets_root already exists in the notebook.
#
# Reads:
#   datasets_root / "gridfm_datakit_data_out" / "case14_ieee" / "raw" / "{bus,branch,gen}_data.parquet"
#
# Percent columns are reported two ways:
# - %_of_records: component-scenario rows, e.g. one bus in one scenario.
# - %_of_cases: unique cases/scenarios with at least one violation.
#
# For branch thermal constraints, rate_a <= 0 is treated as "no thermal rating".

TOL = 1e-9

root_path = datasets_root / "gridfm_datakit_data_out" / "case14_ieee" / "raw"


def read_all_partition_parquets(dataset_dir):
    dataset_dir = Path(dataset_dir)
    parquet_files = sorted(dataset_dir.glob("scenario_partition=*/*.parquet"))

    if not parquet_files:
        raise FileNotFoundError(f"No parquet files found under: {dataset_dir}")

    dfs = []

    for parquet_file in parquet_files:
        df = pd.read_parquet(parquet_file)

        match = re.search(r"scenario_partition=(\d+)", str(parquet_file.parent))
        if match:
            df["scenario_partition"] = int(match.group(1))

        df["source_parquet"] = parquet_file.name
        dfs.append(df)

    return pd.concat(dfs, ignore_index=True)


branch_df = read_all_partition_parquets(root_path / "branch_data.parquet")
bus_df = read_all_partition_parquets(root_path / "bus_data.parquet")
gen_df = read_all_partition_parquets(root_path / "gen_data.parquet")

CASE_COLS = [
    c for c in ["scenario", "load_scenario_idx", "scenario_partition"]
    if c in bus_df.columns
]


def _pct(num, den):
    return 100.0 * num / den if den else 0.0


def _n_cases(df, case_cols):
    if len(df) == 0:
        return 0
    if not case_cols:
        return len(df)
    return df[case_cols].drop_duplicates().shape[0]


def summarize_bound_violations(
    df,
    dataset_name,
    checks,
    id_cols=(),
    case_cols=None,
    active_mask=None,
    tol=TOL,
):
    """
    checks: iterable of (check_name, value_col, min_col, max_col)
            Use None for min_col or max_col when a bound does not exist.
    """
    if case_cols is None:
        case_cols = [
            c for c in ["scenario", "load_scenario_idx", "scenario_partition"]
            if c in df.columns
        ]

    if active_mask is None:
        active_mask = pd.Series(True, index=df.index)
    else:
        active_mask = pd.Series(active_mask, index=df.index).fillna(False).astype(bool)

    d = df.loc[active_mask].copy()

    summaries = []
    details = []
    any_violation = pd.Series(False, index=d.index)

    for check_name, value_col, min_col, max_col in checks:
        needed_cols = [value_col]
        if min_col is not None:
            needed_cols.append(min_col)
        if max_col is not None:
            needed_cols.append(max_col)

        missing_cols = [c for c in needed_cols if c not in d.columns]
        if missing_cols:
            summaries.append({
                "dataset": dataset_name,
                "check": check_name,
                "status": f"skipped; missing columns: {missing_cols}",
            })
            continue

        value = d[value_col]

        lower_applicable = (
            value.notna() & d[min_col].notna()
            if min_col is not None
            else pd.Series(False, index=d.index)
        )
        upper_applicable = (
            value.notna() & d[max_col].notna()
            if max_col is not None
            else pd.Series(False, index=d.index)
        )
        applicable = lower_applicable | upper_applicable

        under = (
            lower_applicable & (value < d[min_col] - tol)
            if min_col is not None
            else pd.Series(False, index=d.index)
        )
        over = (
            upper_applicable & (value > d[max_col] + tol)
            if max_col is not None
            else pd.Series(False, index=d.index)
        )

        violated = under | over
        any_violation = any_violation | violated

        records_checked = int(applicable.sum())
        cases_checked = _n_cases(d.loc[applicable], case_cols)

        summaries.append({
            "dataset": dataset_name,
            "check": check_name,
            "status": "ok",
            "records_checked": records_checked,
            "under_count": int(under.sum()),
            "under_%_of_records": _pct(int(under.sum()), records_checked),
            "over_count": int(over.sum()),
            "over_%_of_records": _pct(int(over.sum()), records_checked),
            "violating_records_total": int(violated.sum()),
            "total_%_of_records": _pct(int(violated.sum()), records_checked),
            "cases_checked": cases_checked,
            "cases_under": _n_cases(d.loc[under], case_cols),
            "under_%_of_cases": _pct(_n_cases(d.loc[under], case_cols), cases_checked),
            "cases_over": _n_cases(d.loc[over], case_cols),
            "over_%_of_cases": _pct(_n_cases(d.loc[over], case_cols), cases_checked),
            "cases_violating_total": _n_cases(d.loc[violated], case_cols),
            "total_%_of_cases": _pct(_n_cases(d.loc[violated], case_cols), cases_checked),
        })

        if violated.any():
            show_cols = []
            for col in [*case_cols, *id_cols, value_col, min_col, max_col]:
                if col is not None and col in d.columns and col not in show_cols:
                    show_cols.append(col)

            detail = d.loc[violated, show_cols].copy()
            detail.insert(0, "dataset", dataset_name)
            detail.insert(1, "check", check_name)

            violation_type = pd.Series("", index=detail.index, dtype="object")
            violation_type.loc[under.loc[violated]] = "under_min"
            violation_type.loc[over.loc[violated]] = "over_max"
            detail.insert(2, "violation_type", violation_type)

            if min_col is not None:
                detail["amount_under"] = (d[min_col] - d[value_col]).where(under).loc[violated]
            if max_col is not None:
                detail["amount_over"] = (d[value_col] - d[max_col]).where(over).loc[violated]

            details.append(detail)

    summary_df = pd.DataFrame(summaries)

    valid_checks = summary_df.get("status", pd.Series(dtype="object")).eq("ok")

    if valid_checks.any():
        all_applicable_rows = pd.Series(False, index=d.index)

        for _, value_col, min_col, max_col in checks:
            if value_col not in d.columns:
                continue

            value = d[value_col]
            applicable = pd.Series(False, index=d.index)

            if min_col is not None and min_col in d.columns:
                applicable = applicable | (value.notna() & d[min_col].notna())
            if max_col is not None and max_col in d.columns:
                applicable = applicable | (value.notna() & d[max_col].notna())

            all_applicable_rows = all_applicable_rows | applicable

        records_checked_any = int(all_applicable_rows.sum())
        cases_checked_any = _n_cases(d.loc[all_applicable_rows], case_cols)

        overall_df = pd.DataFrame([{
            "dataset": dataset_name,
            "records_checked": records_checked_any,
            "records_with_any_violation": int(any_violation.sum()),
            "%_records_with_any_violation": _pct(int(any_violation.sum()), records_checked_any),
            "cases_checked": cases_checked_any,
            "cases_with_any_violation": _n_cases(d.loc[any_violation], case_cols),
            "%_cases_with_any_violation": _pct(
                _n_cases(d.loc[any_violation], case_cols),
                cases_checked_any,
            ),
        }])
    else:
        overall_df = pd.DataFrame([{
            "dataset": dataset_name,
            "records_checked": 0,
            "records_with_any_violation": 0,
            "%_records_with_any_violation": 0.0,
            "cases_checked": 0,
            "cases_with_any_violation": 0,
            "%_cases_with_any_violation": 0.0,
        }])

    detail_df = pd.concat(details, ignore_index=True) if details else pd.DataFrame()
    return summary_df, overall_df, detail_df


# 1) Bus: voltage magnitude limits
bus_summary, bus_overall, bus_violations = summarize_bound_violations(
    bus_df,
    dataset_name="bus",
    checks=[
        ("Vm vs min_vm_pu/max_vm_pu", "Vm", "min_vm_pu", "max_vm_pu"),
    ],
    id_cols=["bus"],
    case_cols=CASE_COLS,
)


# 2) Branch: thermal limits and angle-difference limits
branch_eval = branch_df.copy()

branch_active = (
    branch_eval["br_status"].ne(0)
    if "br_status" in branch_eval.columns
    else pd.Series(True, index=branch_eval.index)
)

# MATPOWER-style convention: rate_a <= 0 means no thermal rating.
if "rate_a" in branch_eval.columns:
    branch_eval["rate_a"] = branch_eval["rate_a"].where(branch_eval["rate_a"] > 0, np.nan)

if {"pf", "qf"}.issubset(branch_eval.columns):
    branch_eval["S_from"] = np.sqrt(branch_eval["pf"] ** 2 + branch_eval["qf"] ** 2)

if {"pt", "qt"}.issubset(branch_eval.columns):
    branch_eval["S_to"] = np.sqrt(branch_eval["pt"] ** 2 + branch_eval["qt"] ** 2)

branch_checks = [
    ("S_from vs rate_a", "S_from", None, "rate_a"),
    ("S_to vs rate_a", "S_to", None, "rate_a"),
]

# Angle-difference check: Va(from_bus) - Va(to_bus), joined from bus_df.
branch_case_cols = [c for c in CASE_COLS if c in branch_eval.columns and c in bus_df.columns]

if (
    {"from_bus", "to_bus", "ang_min", "ang_max"}.issubset(branch_eval.columns)
    and {"bus", "Va"}.issubset(bus_df.columns)
):
    from_va = bus_df[[*branch_case_cols, "bus", "Va"]].rename(
        columns={"bus": "from_bus", "Va": "Va_from"}
    )
    to_va = bus_df[[*branch_case_cols, "bus", "Va"]].rename(
        columns={"bus": "to_bus", "Va": "Va_to"}
    )

    branch_eval = branch_eval.merge(from_va, on=[*branch_case_cols, "from_bus"], how="left")
    branch_eval = branch_eval.merge(to_va, on=[*branch_case_cols, "to_bus"], how="left")
    branch_eval["angle_diff_deg"] = branch_eval["Va_from"] - branch_eval["Va_to"]

    branch_checks.append(
        ("Va_from - Va_to vs ang_min/ang_max", "angle_diff_deg", "ang_min", "ang_max")
    )

branch_summary, branch_overall, branch_violations = summarize_bound_violations(
    branch_eval,
    dataset_name="branch",
    checks=branch_checks,
    id_cols=["idx", "from_bus", "to_bus"],
    case_cols=branch_case_cols,
    active_mask=branch_active,
)


# 3) Gen: active/reactive generator output limits
gen_case_cols = [c for c in CASE_COLS if c in gen_df.columns]

gen_active = (
    gen_df["in_service"].astype(bool)
    if "in_service" in gen_df.columns
    else pd.Series(True, index=gen_df.index)
)

gen_summary, gen_overall, gen_violations = summarize_bound_violations(
    gen_df,
    dataset_name="gen",
    checks=[
        ("p_mw vs min_p_mw/max_p_mw", "p_mw", "min_p_mw", "max_p_mw"),
        ("q_mvar vs min_q_mvar/max_q_mvar", "q_mvar", "min_q_mvar", "max_q_mvar"),
    ],
    id_cols=["idx", "bus"],
    case_cols=gen_case_cols,
    active_mask=gen_active,
)


constraint_summary = pd.concat(
    [bus_summary, branch_summary, gen_summary],
    ignore_index=True,
).round(6)

overall_summary = pd.concat(
    [bus_overall, branch_overall, gen_overall],
    ignore_index=True,
).round(6)

violation_rows = pd.concat(
    [bus_violations, branch_violations, gen_violations],
    ignore_index=True,
)

print("Loaded rows")
display(pd.DataFrame([
    {"dataset": "bus", "rows": len(bus_df), "scenario_partitions": bus_df["scenario_partition"].nunique() if "scenario_partition" in bus_df.columns else None},
    {"dataset": "branch", "rows": len(branch_df), "scenario_partitions": branch_df["scenario_partition"].nunique() if "scenario_partition" in branch_df.columns else None},
    {"dataset": "gen", "rows": len(gen_df), "scenario_partitions": gen_df["scenario_partition"].nunique() if "scenario_partition" in gen_df.columns else None},
]))

print("Per-check constraint violation summary")
display(constraint_summary)

print("Overall summary by dataset")
display(overall_summary)

print("Violation rows, first 50")
display(violation_rows.head(50))
```
